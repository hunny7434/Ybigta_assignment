```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.musinsa.com/app/codimap/lists?style_type=&tag_no=&brand=&display_cnt=60&list_kind=big&sort=comment_cnt&page=1"
res = requests.get(url,headers={'User-agent':'Mozila/5.0'})

soup = BeautifulSoup(res.text, 'html.parser')

codimaps = soup.find_all('div', class_={"style-list-information", "post-information"})
print(codimaps)
```

```python
codimap_category = []
codimap_title = []
codimap_date = []
views = []
comment_numbers = []

for i in range(0, len(codimaps), 2):
    style_info = codimaps[i]
    post_info = codimaps[i + 1]

    category_element = style_info.find('span', class_='style-list-information__text')
    title_element = style_info.find('strong', class_='style-list-information__title')

    if category_element:
        category = category_element.text.strip()
        codimap_category.append(category)
    if title_element:
        title = title_element.text.strip()
        codimap_title.append(title)

    date_element = post_info.find_all('span', class_='post-information__text')[0]
    views_element = post_info.find_all('span', class_='post-information__text')[1]
    comment_element = post_info.find_all('span', class_='post-information__text')[2]

    if date_element:
        date = date_element.text.strip()
        codimap_date.append(date)
    if views_element:
        views_count = views_element.text.split()[1]
        views.append(views_count)
    else:
        views.append('N/A')
    if comment_element:
        comment_count = comment_element.text.split()[1]
        comment_numbers.append(comment_count)
    else:
        comment_numbers.append('N/A')
print(codimap_category)
print(codimap_title)
print(codimap_date)
print(views)
print(comment_numbers)
```

```python
import pandas as pd

df = pd.DataFrame({'codimap_category':codimap_category,
                   'codimap_title':codimap_title,
                   'codimap_date':codimap_date,
                   'views':views,
                   'comment_numbers':comment_numbers})
df.to_csv("test.csv", index = False)
```
